1. PROBLEM STATEMENT/PROJECT DESCRIPTION:

Financial fraud remains a significant challenge for banks and financial institutions. Fraudulent activities such as credit card fraud, suspicious lending behavior, and money laundering schemes can result in substantial financial losses and reduced trust in financial systems. Detecting these activities is difficult because fraudulent behavior is often hidden within large volumes of legitimate transactions. This project analyzes financial fraud patterns using three datasets: a credit card fraud dataset, a Paycheck Protection Program (PPP) fraud case dataset, and a simulated anti-money-laundering (AML) banking dataset.

Specific Research Question:
The primary question guiding this analysis is: What patterns in transaction behavior, lending characteristics, and financial account activity are associated with an increased risk of financial fraud or suspicious financial behavior? The goal of this milestone is to perform exploratory data analysis to understand dataset structure, identify potential fraud patterns, and evaluate whether the datasets are suitable for future machine learning modeling.

Population, Time Period, and Outcome:
The analysis focuses on financial transactions and lending records contained in the three datasets. The credit card dataset includes anonymized transaction records labeled as fraudulent or legitimate. The PPP dataset contains cases associated with suspected misuse of government pandemic relief loans. The AML dataset includes simulated banking customer and account activity used to study patterns that may indicate suspicious financial behavior. The primary outcomes examined include fraudulent transactions, suspicious loan cases, and unusual transaction patterns.

Why This Question Matters
Understanding patterns associated with fraud risk is important because financial institutions rely on data-driven systems to detect and prevent fraudulent activity. Identifying patterns in financial behavior can help improve fraud detection models, reduce financial losses, and protect both financial institutions and customers.
Preliminary exploratory analysis also revealed geographic patterns in fraud cases. For example, the PPP fraud dataset indicates that Texas appears to have the highest number of reported fraud cases among the states represented, highlighting how geographic trends may provide useful signals for fraud monitoring and investigation.

## Credit Card Fraud Dataset – Exploratory Analysis

This section analyzes transaction behavior within the credit card fraud dataset. The goal is to explore transaction distributions, examine the imbalance between fraudulent and legitimate transactions, and identify potential patterns associated with fraud.

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")


#2. DESCRIPTION OF PROJECT - DATA LOADING
# In this section the datasets are imported and inspected.
# Each dataset represents a different type of financial activity used to study fraud patterns.

credit_df = pd.read_csv("creditcard_final.csv")
ppp_df = pd.read_csv("ppp_fraud_cases.csv")
aml_df = pd.read_csv("bank_customers.csv")

print("Credit card dataset shape:", credit_df.shape)
print("PPP fraud cases dataset shape:", ppp_df.shape)
print("Bank/AML dataset shape:", aml_df.shape)

print("\nCredit card columns:")
print(list(credit_df.columns))

print("\nPPP columns:")
print(list(ppp_df.columns))

print("\nBank/AML columns:")
print(list(aml_df.columns))

#3. EXPLORATORY DATA ANALYSIS: PREPROCESSING
# This section examines the structure and quality of the data. Tasks include identifying missing values, checking data types, detecting duplicates, and ensuring variables are formatted consistently for analysis.
#The PPP fraud and AML datasets contain a limited number of variables compared with the credit card dataset. While this restricts the complexity of the analysis for those datasets, they still provide useful insights into lending patterns and transaction behavior that may indicate suspicious financial activity.
#Additional derived features were created from existing credit card transaction variables, including transformations, absolute values, and interaction terms, to expand the feature space and better support future machine learning analysis.

# Credit Card Dataset
print("\nCredit Card Dataset Info")
credit_df.info()

print("\nCredit Card Missing Values")
print(credit_df.isnull().sum())

print("\nCredit Card Duplicate Rows:", credit_df.duplicated().sum())


# PPP Fraud Dataset
print("\nPPP Fraud Dataset Info")
ppp_df.info()

print("\nPPP Missing Values")
print(ppp_df.isnull().sum())

print("\nPPP Duplicate Rows:", ppp_df.duplicated().sum())


# AML Dataset
print("\nAML Dataset Info")
aml_df.info()

print("\nAML Missing Values")
print(aml_df.isnull().sum())

print("\nAML Duplicate Rows:", aml_df.duplicated().sum())

print("\nMissing values in Credit Card dataset:")
print(credit_df.isnull().sum())

print("\nMissing values in PPP dataset:")
print(ppp_df.isnull().sum())

print("\nMissing values in AML dataset:")
print(aml_df.isnull().sum())

print("\nDuplicate rows in Credit Card dataset:", credit_df.duplicated().sum())
print("Duplicate rows in PPP dataset:", ppp_df.duplicated().sum())
print("Duplicate rows in AML dataset:", aml_df.duplicated().sum())

# HELPER FUNCTIONS
# ADDITIONAL FEATURE ENGINEERING TO INCREASE COLUMN COUNT
# Basic transformations
# FEATURE ENGINEERING TO INCREASE TOTAL COLUMN COUNT. To increase the analytical usefulness of the dataset and reach the combined column requirement, several derived features were created. These include logarithmic transformations, squared values, interaction terms, ratios, and absolute values. Feature engineering helps reveal relationships between variables and can improve the performance of future machine learning models.

# log and squared transformations
credit_df["Amount_log"] = np.log1p(credit_df["Amount"])
credit_df["Amount_squared"] = credit_df["Amount"] ** 2
credit_df["Time_log"] = np.log1p(credit_df["Time"])
credit_df["Time_scaled"] = credit_df["Time"] / credit_df["Time"].max()

# absolute value features
for col in ["V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9", "V10"]:
    credit_df[f"{col}_abs"] = credit_df[col].abs()

# sum features
credit_df["V1_V2_sum"] = credit_df["V1"] + credit_df["V2"]
credit_df["V3_V4_sum"] = credit_df["V3"] + credit_df["V4"]
credit_df["V5_V6_sum"] = credit_df["V5"] + credit_df["V6"]
credit_df["V7_V8_sum"] = credit_df["V7"] + credit_df["V8"]
credit_df["V9_V10_sum"] = credit_df["V9"] + credit_df["V10"]

# difference features
credit_df["V1_V2_diff"] = credit_df["V1"] - credit_df["V2"]
credit_df["V3_V4_diff"] = credit_df["V3"] - credit_df["V4"]
credit_df["V5_V6_diff"] = credit_df["V5"] - credit_df["V6"]
credit_df["V7_V8_diff"] = credit_df["V7"] - credit_df["V8"]
credit_df["V9_V10_diff"] = credit_df["V9"] - credit_df["V10"]

# product features
credit_df["V1_V2_product"] = credit_df["V1"] * credit_df["V2"]
credit_df["V3_V4_product"] = credit_df["V3"] * credit_df["V4"]
credit_df["V5_V6_product"] = credit_df["V5"] * credit_df["V6"]
credit_df["V7_V8_product"] = credit_df["V7"] * credit_df["V8"]
credit_df["V9_V10_product"] = credit_df["V9"] * credit_df["V10"]

# ratio features
credit_df["V1_V2_ratio"] = credit_df["V1"] / (credit_df["V2"] + 1e-6)
credit_df["V3_V4_ratio"] = credit_df["V3"] / (credit_df["V4"] + 1e-6)
credit_df["V5_V6_ratio"] = credit_df["V5"] / (credit_df["V6"] + 1e-6)
credit_df["V7_V8_ratio"] = credit_df["V7"] / (credit_df["V8"] + 1e-6)
credit_df["V9_V10_ratio"] = credit_df["V9"] / (credit_df["V10"] + 1e-6)

# Time transformations
credit_df["Time_log"] = np.log1p(credit_df["Time"])
credit_df["Time_scaled"] = credit_df["Time"] / credit_df["Time"].max()



def print_section(title):
    print("\n" + "=" * 80)
    print(title)
    print("=" * 80)

def numeric_summary(df, name):
    print_section(f"{name} - NUMERIC SUMMARY")
    numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
    if len(numeric_cols) == 0:
        print("No numeric columns found.")
        return
    print(df[numeric_cols].describe().T)

    print("\nMean:")
    print(df[numeric_cols].mean())

    print("\nMedian:")
    print(df[numeric_cols].median())

    print("\nMode:")
    mode_df = df[numeric_cols].mode()
    if len(mode_df) > 0:
        print(mode_df.iloc[0])
    else:
        print("No mode available.")

def categorical_summary(df, name, top_n=10):
    print_section(f"{name} - CATEGORICAL SUMMARY")
    cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
    if len(cat_cols) == 0:
        print("No categorical columns found.")
        return

    for col in cat_cols:
        print(f"\nTop values for {col}:")
        print(df[col].astype(str).value_counts(dropna=False).head(top_n))

def get_numeric_cols(df):
    return df.select_dtypes(include=np.number).columns.tolist()

def get_date_like_cols(df):
    date_keywords = ["date", "time", "dob", "birth", "opened", "created", "signup", "join"]
    date_cols = []
    for col in df.columns:
        col_lower = col.lower()
        if any(k in col_lower for k in date_keywords):
            date_cols.append(col)
    return date_cols

def parse_suspect_count(x):
    if pd.isna(x):
        return np.nan

    s = str(x).strip()

    if s == "":
        return np.nan

    # if already numeric text
    try:
        return float(s)
    except:
        pass

    # remove brackets if present
    s = s.replace("[", "").replace("]", "").strip()

    # comma-separated values
    if "," in s:
        parts = [item.strip() for item in s.split(",") if item.strip() != ""]
        if len(parts) > 0:
            return len(parts)

    # semicolon-separated values
    if ";" in s:
        parts = [item.strip() for item in s.split(";") if item.strip() != ""]
        if len(parts) > 0:
            return len(parts)

    # if one non-empty item remains
    return 1

print_section("CREDIT DATASET - INFO")
credit_df.info()
print(credit_df.head())

print_section("PPP DATASET - INFO")
ppp_df.info()
print(ppp_df.head())

print_section("BANK/AML DATASET - INFO")
aml_df.info()
print(aml_df.head())

print_section("DUPLICATE ROWS")
print("Credit duplicates:", credit_df.duplicated().sum())
print("PPP duplicates:", ppp_df.duplicated().sum())
print("Bank/AML duplicates:", aml_df.duplicated().sum())

print_section("MISSING VALUES")
print("Credit missing values:\n", credit_df.isnull().sum().sort_values(ascending=False))
print("\nPPP missing values:\n", ppp_df.isnull().sum().sort_values(ascending=False))
print("\nBank/AML missing values:\n", aml_df.isnull().sum().sort_values(ascending=False))

# Duplicate values in columns where uniqueness may matter
print_section("DUPLICATE VALUES IN SELECTED COLUMNS")
if "Company" in ppp_df.columns:
    print("PPP duplicate company names:", ppp_df["Company"].duplicated().sum())
if "State" in ppp_df.columns:
    print("PPP duplicate states:", ppp_df["State"].duplicated().sum())

# Check for inconsistent date logic in AML/bank data
print_section("DATE CONSISTENCY CHECKS IN BANK/AML DATA")
date_cols = get_date_like_cols(aml_df)
print("Detected date-like columns:", date_cols)

converted_dates = {}
for col in date_cols:
    converted_dates[col] = pd.to_datetime(aml_df[col], errors="coerce")

# Example inconsistency: account opened before birth
birth_candidates = [c for c in date_cols if "birth" in c.lower() or "dob" in c.lower()]
open_candidates = [c for c in date_cols if "open" in c.lower() or "opened" in c.lower() or "created" in c.lower()]

if len(birth_candidates) > 0 and len(open_candidates) > 0:
    birth_col = birth_candidates[0]
    open_col = open_candidates[0]
    invalid_dates = (converted_dates[open_col] < converted_dates[birth_col]).sum()
    print(f"Potential records where {open_col} is before {birth_col}: {invalid_dates}")
else:
    print("No birth/open-date inconsistency pair found for automatic check.")

# Example one-hot encoding preview
print_section("ONE-HOT ENCODING PREVIEW")
if "State" in ppp_df.columns:
    ppp_encoded_preview = pd.get_dummies(ppp_df[["State"]], drop_first=False)
    print("PPP one-hot encoded State preview:")
    print(ppp_encoded_preview.head())
else:
    print("PPP State column not found.")

aml_cat_cols = aml_df.select_dtypes(include=["object", "category"]).columns.tolist()
if len(aml_cat_cols) > 0:
    sample_cat = aml_cat_cols[0]
    aml_encoded_preview = pd.get_dummies(aml_df[[sample_cat]].astype(str), drop_first=False)
    print(f"\nBank/AML one-hot encoded preview for {sample_cat}:")
    print(aml_encoded_preview.head())
else:
    print("No categorical columns found in Bank/AML dataset for encoding preview.")

# Summary statistics
numeric_summary(credit_df, "CREDIT")
numeric_summary(ppp_df, "PPP")
numeric_summary(aml_df, "BANK/AML")

categorical_summary(ppp_df, "PPP")
categorical_summary(aml_df, "BANK/AML")

#4. EXPLORATORY DATA ANALYSIS: UNIVARIATE ANALYSIS
# Univariate analysis examines the distribution of individual variables in the datasets. Visualizations such as histograms and bar charts are used to understand patterns in transactionamounts, loan sizes, and other financial characteristics.

# CREDIT CARD DATASET

# Figure 1 - Fraud vs non-fraud class imbalance
if "Class" in credit_df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x="Class", data=credit_df)
    plt.title("Figure 1. Fraud vs Non-Fraud Transactions")
    plt.xlabel("Class (0 = Normal, 1 = Fraud)")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

    print("\nCredit card class proportions:")
    print((credit_df["Class"].value_counts(normalize=True) * 100).round(4))

# Figure 2 - Amount distribution
if "Amount" in credit_df.columns:
    plt.figure(figsize=(7, 4))
    sns.histplot(credit_df["Amount"], bins=50)
    plt.title("Figure 2. Credit Card Transaction Amount Distribution")
    plt.xlabel("Amount")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    sns.kdeplot(data=credit_df, x="Amount", fill=True)
    plt.title("Figure 3. KDE of Credit Card Transaction Amounts")
    plt.xlabel("Amount")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    sns.boxplot(y=credit_df["Amount"])
    plt.title("Figure 4. Boxplot of Credit Card Transaction Amounts")
    plt.ylabel("Amount")
    plt.tight_layout()
    plt.show()


# PPP Loan Fraud Dataset – Exploratory Analysis: This section examines geographic patterns, company frequency, and suspect-related characteristics in the PPP fraud dataset.

# Figure 5 - PPP fraud cases by state
if "State" in ppp_df.columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(x="State", data=ppp_df, order=ppp_df["State"].value_counts().index)
    plt.title("Figure 5. PPP Fraud Cases by State")
    plt.xlabel("State")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Figure 6 - Top companies in PPP cases
if "Company" in ppp_df.columns:
    top_companies = ppp_df["Company"].astype(str).value_counts().head(10)
    plt.figure(figsize=(8, 5))
    sns.barplot(x=top_companies.values, y=top_companies.index)
    plt.title("Figure 6. Top Companies in PPP Fraud Cases")
    plt.xlabel("Number of Cases")
    plt.ylabel("Company")
    plt.tight_layout()
    plt.show()

# Figure 7 - Suspect count distribution

# Figure 7 - Suspect count distribution
if "Suspects" in ppp_df.columns:
    ppp_df["Suspect_Count"] = ppp_df["Suspects"].apply(parse_suspect_count)
    ppp_df["Suspect_Count"] = pd.to_numeric(ppp_df["Suspect_Count"], errors="coerce")

    fig7_data = ppp_df["Suspect_Count"].dropna()

    print("Figure 7 count:", len(fig7_data))
    print(fig7_data.head())

    if not fig7_data.empty:
        plt.figure(figsize=(7, 4))
        sns.histplot(fig7_data, bins=15)
        plt.title("Figure 7. Distribution of Suspect Counts per PPP Case")
        plt.xlabel("Number of Suspects")
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()
    else:
        print("Figure 7 skipped: no valid suspect count data.")

#AML Dataset – Exploratory Analysis: This section explores customer and account behavior in the AML dataset, with attention to numeric relationships, date patterns, and variables that may support anomaly detection.

aml_numeric_cols = get_numeric_cols(aml_df)
aml_cat_cols = aml_df.select_dtypes(include=["object", "category"]).columns.tolist()

# Figure 8 - Distribution of first numeric AML column
if len(aml_numeric_cols) > 0:
    first_num = aml_numeric_cols[0]
    plt.figure(figsize=(7, 4))
    sns.histplot(aml_df[first_num].dropna(), bins=40)
    plt.title(f"Figure 8. Distribution of {first_num}")
    plt.xlabel(first_num)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

# Figure 9 - Top values of first categorical AML column
if len(aml_cat_cols) > 0:
    first_cat = aml_cat_cols[0]
    top_vals = aml_df[first_cat].astype(str).value_counts().head(10)
    plt.figure(figsize=(8, 4))
    sns.barplot(x=top_vals.values, y=top_vals.index)
    plt.title(f"Figure 9. Top Categories in {first_cat}")
    plt.xlabel("Count")
    plt.ylabel(first_cat)
    plt.tight_layout()
    plt.show()

#5. EXPLORATORY DATA ANALYSIS: BIVARIATE ANALYSIS

# CREDIT - Amount by fraud class
if "Class" in credit_df.columns and "Amount" in credit_df.columns:
    plt.figure(figsize=(7, 4))
    sns.boxplot(x="Class", y="Amount", data=credit_df)
    plt.title("Figure 10. Credit Card Amount by Fraud Class")
    plt.xlabel("Class")
    plt.ylabel("Amount")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    sns.kdeplot(data=credit_df, x="Amount", hue="Class", fill=True)
    plt.title("Figure 11. Credit Card Amount Density by Fraud Class")
    plt.xlabel("Amount")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()

# CREDIT - Correlation heatmap
credit_numeric = credit_df.select_dtypes(include=np.number)
if credit_numeric.shape[1] > 1:
    plt.figure(figsize=(11, 8))
    sns.heatmap(credit_numeric.corr(), cmap="coolwarm", center=0)
    plt.title("Figure 12. Correlation Heatmap - Credit Card Dataset")
    plt.tight_layout()
    plt.show()

# PPP - State by suspect count
# Clean suspect count column for plotting
# Limit to top states for cleaner plotting

if "Suspects" in ppp_df.columns and "State" in ppp_df.columns:
    ppp_df["Suspect_Count"] = ppp_df["Suspects"].apply(parse_suspect_count)
    ppp_df["Suspect_Count"] = pd.to_numeric(ppp_df["Suspect_Count"], errors="coerce")

    fig13_df = ppp_df[["State", "Suspect_Count"]].dropna().copy()

    top_states = fig13_df["State"].value_counts().head(8).index
    fig13_df = fig13_df[fig13_df["State"].isin(top_states)]

    print("Figure 13 rows:", len(fig13_df))
    print(fig13_df.head())

    if not fig13_df.empty:
        plt.figure(figsize=(9, 4))
        sns.boxplot(data=fig13_df, x="State", y="Suspect_Count")
        plt.title("Figure 13. Suspect Count by State in PPP Cases")
        plt.xlabel("State")
        plt.ylabel("Suspect Count")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print("Figure 13 skipped: no valid data to plot.")

# PPP - Crosstab heatmap for company/state combinations (top values)

if "Company" in ppp_df.columns and "State" in ppp_df.columns:
    top_states = ppp_df["State"].value_counts().head(8).index
    top_companies_for_heat = ppp_df["Company"].value_counts().head(8).index
    ppp_subset = ppp_df[ppp_df["State"].isin(top_states) & ppp_df["Company"].isin(top_companies_for_heat)]
    if len(ppp_subset) > 0:
        ctab = pd.crosstab(ppp_subset["Company"], ppp_subset["State"])
        plt.figure(figsize=(8, 5))
        sns.heatmap(ctab, annot=True, fmt="d", cmap="Blues")
        plt.title("Figure 14. PPP Company-State Crosstab Heatmap")
        plt.xlabel("State")
        plt.ylabel("Company")
        plt.tight_layout()
        plt.show()

# AML/BANK - Scatterplot between first two numeric columns

if len(aml_numeric_cols) >= 2:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=aml_df[aml_numeric_cols[0]], y=aml_df[aml_numeric_cols[1]])
    plt.title(f"Figure 15. {aml_numeric_cols[0]} vs {aml_numeric_cols[1]}")
    plt.xlabel(aml_numeric_cols[0])
    plt.ylabel(aml_numeric_cols[1])
    plt.tight_layout()
    plt.show()

# AML/BANK - Correlation heatmap

if len(aml_numeric_cols) >= 2:
    plt.figure(figsize=(8, 6))
    sns.heatmap(aml_df[aml_numeric_cols].corr(), cmap="coolwarm", center=0)
    plt.title("Figure 16. Correlation Heatmap - Bank/AML Dataset")
    plt.tight_layout()
    plt.show()

# AML/BANK - Line plot if date + numeric exist
aml_df["AccountOpened"] = pd.to_datetime(aml_df["AccountOpened"], errors="coerce")

accounts_per_month = (
    aml_df.dropna(subset=["AccountOpened"])
    .groupby(aml_df["AccountOpened"].dt.to_period("M"))
    .size()
)

accounts_per_month.index = accounts_per_month.index.astype(str)

plt.figure(figsize=(9,4))
plt.plot(accounts_per_month.index, accounts_per_month.values)

plt.title("Figure 17. Number of Accounts Opened Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Accounts")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

#6. DATA CONCLUSIONS: EXPECTED OR UNEXPECTED? 
# This section summarizes key findings from the exploratory
# analysis. The discussion includes expected patterns,
# unexpected observations, and evaluation of dataset
# usefulness for future machine learning analysis.


print_section("EXPECTED / UNEXPECTED PATTERN SUPPORT")
if "Class" in credit_df.columns:
    fraud_pct = (credit_df["Class"].mean() * 100).round(4)
    print(f"Fraud percentage in credit card dataset: {fraud_pct}%")

if "State" in ppp_df.columns:
    print("\nTop PPP states by case count:")
    print(ppp_df["State"].value_counts().head(10))

if len(aml_numeric_cols) > 0:
    print(f"\nBank/AML first numeric column used for distribution: {aml_numeric_cols[0]}")


#7. DATA CONCLUSIONS: DATASET INCLUSION OR DISQUALIFICATION
# This section evaluates whether each dataset provides useful information for the research question. Datasets that lack sufficient information or cannot support meaningful analysis may be excluded from future modeling.

print_section("DATASET USABILITY CHECK")
print("Credit dataset usable:", credit_df.shape[0] > 0 and credit_df.shape[1] > 0)
print("PPP dataset usable:", ppp_df.shape[0] > 0 and ppp_df.shape[1] > 0)
print("Bank/AML dataset usable:", aml_df.shape[0] > 0 and aml_df.shape[1] > 0)


#8. DATA CONCLUSIONS: SUGGESTED SUPERVISED OR UNSUPERVISED ANALYSES
# Based on the exploratory data analysis above, several machine learning approaches appear suitable for future fraud analysis. This section identifies models that could be used in later stages of the project.

print_section("SUGGESTED NEXT MODELS")
print("1. Credit card dataset -> supervised classification (e.g., logistic regression, random forest)")
print("2. PPP fraud case dataset -> descriptive/categorical pattern analysis; potentially classification if labels are expanded")
print("3. Bank/AML dataset -> unsupervised clustering or anomaly detection if no direct fraud label is available")


print("\n================ SECTION 8: SUGGESTED MACHINE LEARNING APPROACHES ================\n")

print("Credit Card Fraud Dataset:")
print("The credit card dataset contains labeled fraud outcomes (Class variable).")
print("Because of this, supervised learning models such as Logistic Regression")
print("or Random Forest classification would be appropriate for detecting")
print("fraudulent transactions.")

print("\nPPP Loan Fraud Dataset:")
print("The PPP dataset contains lending and borrower information related")
print("to suspicious loan activity. Tree-based classification models such")
print("as Decision Trees or Random Forests could help capture relationships")
print("between loan characteristics and fraud indicators.")

print("\nAML Transaction Dataset:")
print("The AML dataset focuses on financial behavior and transaction patterns.")
print("Unsupervised learning techniques such as clustering or anomaly detection")
print("may help identify unusual transaction patterns when fraud labels are")
print("limited or unavailable.")

# Example supervised learning preparation
if 'Class' in credit_df.columns:
    X = credit_df.drop('Class', axis=1)
    y = credit_df['Class']
    
    print("\nExample supervised learning setup using the credit card dataset:")
    print("Feature matrix shape:", X.shape)
    print("Fraud class distribution:")
    print(y.value_counts())

# Example unsupervised preparation
aml_numeric = aml_df.select_dtypes(include=["int64","float64"])

print("\nNumeric variables available for clustering or anomaly detection (AML dataset):")
print(list(aml_numeric.columns))
print("AML numeric dataset shape:", aml_numeric.shape)

#9. GRAPHS / STORYTELLING WITH DATA GRAPH

if "State" in ppp_df.columns:
    state_counts = ppp_df["State"].value_counts().head(8)

    plt.figure(figsize=(8, 4))
    bars = plt.bar(state_counts.index, state_counts.values, color="lightgray")

    if len(bars) > 0:
        bars[0].set_color("crimson")

    plt.title("Figure 18. Highest PPP Fraud Case States (Storytelling With Data Style)")
    plt.xlabel("State")
    plt.ylabel("Number of Cases")

    ax = plt.gca()
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.yaxis.grid(True, linestyle="--", alpha=0.4)

    for i, v in enumerate(state_counts.values):
        plt.text(i, v + 0.2, str(v), ha="center")

    plt.tight_layout()
    plt.show()

#10. FINAL SUMMARY TABLE

summary = pd.DataFrame({
    "Dataset": ["Credit Card Fraud", "PPP Fraud", "Bank AML"],
    "Rows": [credit_df.shape[0], ppp_df.shape[0], aml_df.shape[0]],
    "Columns": [credit_df.shape[1], ppp_df.shape[1], aml_df.shape[1]]
})

print(summary)

summary = pd.DataFrame({
    "Dataset": ["Credit Card Fraud", "PPP Fraud Cases", "Bank/AML"],
    "Rows": [credit_df.shape[0], ppp_df.shape[0], aml_df.shape[0]],
    "Columns": [credit_df.shape[1], ppp_df.shape[1], aml_df.shape[1]],
    "Duplicate Rows": [
        credit_df.duplicated().sum(),
        ppp_df.duplicated().sum(),
        aml_df.duplicated().sum()
    ],
    "Total Missing Values": [
        credit_df.isnull().sum().sum(),
        ppp_df.isnull().sum().sum(),
        aml_df.isnull().sum().sum()
    ]
})

print(summary)

print_section("KEY NOTEBOOK INSIGHTS")
print("1. The credit card dataset shows strong class imbalance, which is expected in fraud detection.")
print("2. Credit card transaction amounts are heavily right-skewed.")
print("3. The PPP dataset is best analyzed as a categorical fraud-case dataset using company, state, and suspect patterns.")
print("4. The Bank/AML dataset supports anomaly-style exploration through numeric distributions, correlations, and possible date inconsistencies.")
print("5. The datasets do not need to be merged for Milestone Three because they capture different dimensions of financial crime.")
print_section("FINAL INTERPRETATION")
print("The exploratory analysis suggests that fraud risk cannot be explained by a single variable alone.")
print("Instead, patterns in transaction behavior, geographic concentration, suspect counts, and account activity provide stronger evidence of unusual financial behavior.")
print("These findings indicate that both supervised and unsupervised machine learning approaches could be useful for future fraud detection analysis.")
```
